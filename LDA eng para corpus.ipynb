```python
from nltk.tokenize import RegexpTokenizer
from stop_words import get_stop_words
from nltk.stem.porter import PorterStemmer
from gensim import corpora, models
import re, pprint
import gensim
import os
from urllib import request
```

```python
file1 = '/home/ykim/nltk_data/corpora/genesis/english-web.txt'
```

```python
book1 = open(file1, 'r').read()
doc_set = [book1]
tokenizer = RegexpTokenizer(r'\w+')
en_stop = get_stop_words('en')
p_stemmer = PorterStemmer()
```

```python
corpus
```

```python
#create corpus composed of paragraphs as docs
for i in doc_set:
    raw = i.lower()
    #texts = tokenizer.tokenize(raw)
    docs = re.split('[\n\t]+', raw)
exclude = [' '];
corpus = [doc for doc in docs if doc not in exclude]
```

```python
new_corpus = []
for i in corpus:
    tokens = tokenizer.tokenize(i)
    #print(tokens)
    
    stopped_tokens = [i for i in tokens if not i in en_stop]
    stemmed_tokens = [p_stemmer.stem(i) for i in stopped_tokens]
        
    #stemmed_tokens = lemmatizer.lemmatize(stopped_tokens)
    
    new_corpus.append(stemmed_tokens)
    #print(stemmed_tokens)
```

```python
#new_corpus
```

```python
dictionary = corpora.Dictionary(new_corpus)
```

```python
#dictionary.filter_n_most_frequent(200)
```

```python
bags = [dictionary.doc2bow(doc) for doc in new_corpus]
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(bags, num_topics=10, id2word = dictionary, passes=40)
```

```python
print(ldamodel.print_topics(num_topics=10, num_words=3))
```
