```python
from nltk.tokenize import RegexpTokenizer
from stop_words import get_stop_words
from nltk.stem.porter import PorterStemmer
from gensim import corpora, models
import gensim
import os
```

```python
file1 = '/home/ykim/nltk_data/corpora/genesis/english-web.txt'
file2 = '/home/ykim/nltk_data/corpora/gutenberg/bible-kjv.txt'
```

```python
book1 = open(file1, 'r').read()
book2 = open(file2, 'r').read()
```

```python
doc_set = [book1,book2]
```

```python
tokenizer = RegexpTokenizer(r'\w+')
```

```python
en_stop = get_stop_words('en')
```

```python
p_stemmer = PorterStemmer()
```

```python
texts = []
```

```python
for i in doc_set:
    raw = i.lower()
    tokens = tokenizer.tokenize(raw)
    
    stopped_tokens = [i for i in tokens if not i in en_stop]
    stemmed_tokens = [p_stemmer.stem(i) for i in stopped_tokens]
    
    texts.append(stopped_tokens)
    
```

```python
texts[0]
```

```python
len(texts[0])
```

```python
dictionary = corpora.Dictionary(texts)
```

```python
#print(dictionary.token2id)
```

```python
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=3, id2word = dictionary, passes=40)
```

```python
print(ldamodel.print_topics(num_topics=3, num_words=5))
```
